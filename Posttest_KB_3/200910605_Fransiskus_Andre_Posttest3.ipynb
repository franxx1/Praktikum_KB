# Imort Library

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

# Mengakses file

```python
df = pd.read_csv("Breast_Cancer.csv")
```

# Membuat Grafik Batang

```python

plt.bar((df["6th Stage"]), (df["Tumor Size"]), color = "#D58BDD", align = "edge")
plt.title("Perbandingan jumlah Tingkat Stadium dan Ukuran Tumor")
plt.xlabel("Tingkat Stadium")
plt.ylabel("Ukuran Tumor")
plt.show()
```

diagram batang diatas menunjukan perbandingan jumlah tingkat stadium tumor dan ukuran tumor, dapat kita lihat bahwa tingkat stadium IIIA,IIIC, dan IIIB, memiliki jumlah terbanyak dibandingka tingkat stadium tumor yang lain

# Membuat grafik heatmap

```python
sns.heatmap(df.corr(), cmap = "Blues", annot = True)
```

Grafik diatas menunjukan korelasi antar atribut didalam data yang ada, dimana jika warna semakin gelap maka dapat disimpulkan korelasi antar 2 atribut tersebut dapat dikatakan baik, seperti korelasi anar Tumor size dan Reginol Node Positive. sebagai catat untuk angka diagonal itu tidak termasuk kedalam korelasi, karena atribut tersebut berkolerasi dengan dirinya sendiri

# Membuat grafik pie

```python
df_stage = df.dropna(subset = ['6th Stage'])
df_pie = df_stage['6th Stage'].value_counts().rename_axis("Stage").reset_index(name = 'Category')
plt.figure(figsize=(10,10))
plt.pie(df_pie.Category, labels=df_pie.Stage, startangle=90,
autopct='%.1f%%')
plt.title('Stage')
plt.show()
```

Grafik pie diatas menujukan perbadingan tiap data pada kolom 6t Stage, dimana data dengan labelIIB memiliki jumlah yang banyak dari data lainya
