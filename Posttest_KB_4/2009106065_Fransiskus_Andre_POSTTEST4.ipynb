# DATA CLEANING

```python
df.iloc[90:1200,12:15] = np.nan
```

### Mengapus Nilai NULL

```python
imputer_mean = SimpleImputer(strategy= "mean")
imputer_modus = SimpleImputer(strategy = "most_frequent")
imputer_median  = SimpleImputer(strategy = "median")
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
df["Regional Node Examined"] = imputer_mean.fit_transform(df[["Regional Node Examined"]])
df["Reginol Node Positive"] = imputer_modus.fit_transform(df[["Reginol Node Positive"]])
df["Survival Months"] = imputer_median.fit_transform(df[["Survival Months"]])

```

```python
df.isna().sum()
```

### Menghapus Nilai Duplkat

```python
df[df.duplicated()]
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```
